```python
import numpy as np
def zigmoid(z):
  return 1/(1+np.exp(-z))

np.random.seed(42)
X = np.random.rand(100)
y = 2 + 3*X + np.random.randn(100)*0.1
m = np.zeros_like(X[0])
b = 0
n = len(X)
lr = 0.0001
epochs = 50
for epoch in range(1,epochs+1):
  for i in range(len(X)):
    z = np.dot(m,X[i])+b
    y_pred = zigmoid(z)
    gr_wrt_m = X[i] * (y[i] - y_pred)
    gr_wrt_b = y[i]-y_pred
    print(f"Epoch {epoch}, Data Point {i}: Gradient w.r.t m: {gr_wrt_m}, Gradient w.r.tc: {gr_wrt_b}")
    m += lr *gr_wrt_m
    b += lr*gr_wrt_b
      
predictions=[]
for i in range(len(X)):
  z = np.dot(m,X[i])+b
  y_pred = zigmoid(z)
  if y_pred>=0.5:
    predictions.append(1)
  else:
    predictions.append(0)
print(m)
print(b)
```
