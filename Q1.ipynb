```python
import numpy as np
import matplotlib.pyplot as plt

def gradient_descent(X, y, learning_rate, epochs):
    m = 0  # Initial slope
    b = 0  # Initial intercept
    n = len(X)  # Number of data points
    costs = []  # To store the cost at each epoch

    for epoch in range(epochs):
        y_pred = m * X + b  # Predicted values
        error = y_pred - y  # Error between predicted and actual values

        # Update parameters
        m -= learning_rate * (1/n) * np.sum(X * error)
        b -= learning_rate * (1/n) * np.sum(error)

        # Calculate and store cost every 100 epochs
        if epoch % 100 == 0:
            cost = (1/(2*n)) * np.sum(error**2)
            costs.append(cost)
            print(f'Epoch {epoch}, Cost: {cost}')

    # Plot epoch vs. cost
    plt.plot(range(0, epochs, 100), costs, marker='o')
    plt.xlabel('Epoch')
    plt.ylabel('Cost')
    plt.title('Epoch vs. Cost')
    plt.show()

    return m, b

# Generate some random data for demonstration
np.random.seed(42)
X = np.random.rand(100)
y = 2 + 3*X + np.random.randn(100) * 0.1

learning_rate = 0.11
epochs = 1000
m, b = gradient_descent(X, y, learning_rate, epochs)

print("Slope (m):", m)
print("Intercept (b):", b)
```
